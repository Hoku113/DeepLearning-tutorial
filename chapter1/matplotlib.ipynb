# This is a tutorial matplotlib python code

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create data
x = np.arange(0, 6, 0.1) # Display from 0 to 6 in 0.1 increments
y = np.sin(x)

# draw a sin graph
plt.plot(x, y)
plt.show()
```

```python
# draw a cos graph
y2 = np.cos(x)

plt.plot(x, y, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x") # x label
plt.ylabel("y") # y label
plt.title("sin & cos") # graph title
plt.legend()
plt.show()
```
